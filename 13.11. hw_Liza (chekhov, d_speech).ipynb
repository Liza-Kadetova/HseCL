(```re.findall```+ работа с файлами) Детектим прямую речь в тексте. В файле **d_speech.txt** лежит текст, в котором есть строки с прямой речью. Напишите регулярное выражение, которое вытащит все строки с прямой речью. Запишите их (питоном, не руками) в отдельный .txt файл.



````
import re
with open('d_speech.txt', 'r') as myfile:
  myfile=myfile.read()
speech = re.findall(r'— .+?—|— .+|«.+?—', myfile) #получаем список
print(speech)
````

````
type(speech)
````

````
myfile_new = open('d_speech3.txt', 'w')
print(type( myfile_new))
````

````
for n in speech:
  myfile_new.write(n.strip('—,«,»'))
myfile_new.close()
myfile_new = open('d_speech3.txt', 'r')
myfile_new.read()
````
