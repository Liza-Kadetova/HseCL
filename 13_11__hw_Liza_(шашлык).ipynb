(```re.sub```) В документе **ея.txt** замените все случаи употребления строки "язык" на "шашлык"

````
import re
````

````
eya = open('ея.txt', 'r')
text = eya.read()
print(text)
eya.close()
````

````
text2 = re.sub(r"(Я|я)з(ы|ы́)к", r"шашлык", text) #но язык с заклавной будет заменять на шашлык со строчной :(
print(text2)
````

````
eya2 = open('eya2.txt', 'w')
eya2.write(text2)
eya2.close()
````

````
eya2 = open('eya2.txt', 'r')
text3 = eya2.read()
print(text3)
````
